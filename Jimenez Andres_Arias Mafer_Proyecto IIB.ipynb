```python
import os
import json
import faiss
import torch
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
from torchvision import transforms
from transformers import CLIPProcessor, CLIPModel
from sentence_transformers import SentenceTransformer
import shutil

# Configurar dispositivo
device = "cuda" if torch.cuda.is_available() else "cpu"

# Modelos
image_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
image_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
text_model = SentenceTransformer("clip-ViT-B-32")

# Dataset
df = pd.read_csv("/content/fashion_dataset/styles.csv", on_bad_lines='skip')
df = df.dropna(subset=['id', 'gender', 'masterCategory', 'subCategory', 'articleType', 'baseColour', 'usage', 'productDisplayName'])
df['id'] = df['id'].astype(int)
df = df.head(5000)

# Rutas
image_dir = "/content/fashion_dataset/images"
save_path = "/content/drive/MyDrive/fashion_search_1"
os.makedirs(save_path, exist_ok=True)

# Embeddings
image_embeddings = []
text_descriptions = []
metadata = []

print("🔄 Procesando imágenes con CLIP...")
for idx in tqdm(df['id']):
    img_path = os.path.join(image_dir, f"{idx}.jpg")
    if os.path.exists(img_path):
        try:
            image_pil = Image.open(img_path).convert("RGB")
            image_tensor = image_processor(images=image_pil, return_tensors="pt").to(device)
            with torch.no_grad():
                img_emb = image_model.get_image_features(**image_tensor).cpu().numpy().flatten()
            image_embeddings.append(img_emb)

            row = df.loc[df['id'] == idx].iloc[0]
            description = f"{row['productDisplayName']} ({row['articleType']}, Color: {row['baseColour']})"
            text_descriptions.append(description)
            metadata.append({
                "id": int(idx),
                "path": img_path,
                "description": description
            })

        except Exception as e:
            print(f"❌ Error con imagen {idx}: {e}")
            continue

# Validación
if len(image_embeddings) == 0:
    raise ValueError("No se generaron embeddings de imagen.")

# Guardar metadata
with open(os.path.join(save_path, "clip_metadata.json"), "w") as f:
    json.dump(metadata, f)

# 👉 Guardar índice de imágenes
image_embeddings = np.array(image_embeddings).astype('float32')
faiss_index_img = faiss.IndexFlatL2(image_embeddings.shape[1])
faiss_index_img.add(image_embeddings)
faiss.write_index(faiss_index_img, os.path.join(save_path, "clip_image_index.faiss"))

# 👉 Generar y guardar embeddings de texto
print("💬 Codificando descripciones con CLIP text encoder...")
text_embeddings = text_model.encode(text_descriptions, convert_to_numpy=True)
faiss_index_txt = faiss.IndexFlatL2(text_embeddings.shape[1])
faiss_index_txt.add(text_embeddings)
faiss.write_index(faiss_index_txt, os.path.join(save_path, "fashion_text_index.faiss"))
np.save(os.path.join(save_path, "fashion_text_embeddings.npy"), text_embeddings)

print("✅ Todo guardado en Google Drive:", save_path)
```

```python


import gradio as gr
import torch
import faiss
import numpy as np
from PIL import Image
import json
import os
import google.generativeai as genai
from transformers import CLIPProcessor, CLIPModel

# Configurar Gemini API - CAMBIA POR TU API KEY
os.environ["GOOGLE_API_KEY"] = "AIzaSyAXluEbiV8lp0TJzN0s6Av8qHWDDFEMZag"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# Instanciar el modelo Gemini
modelo_gemini = genai.GenerativeModel("gemini-1.5-flash")

# Cargar modelo CLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Ruta base en Drive
base_path = "/content/drive/MyDrive/fashion_search_1"

# Cargar índices
image_index = faiss.read_index(f"{base_path}/clip_image_index.faiss")
text_index = faiss.read_index(f"{base_path}/fashion_text_index.faiss")

# Cargar metadatos
with open(f"{base_path}/clip_metadata.json", "r") as f:
    metadata = json.load(f)

metadata = metadata[:5000]

def preprocess_image(image):
    """Preprocesar imagen para obtener embeddings con CLIP"""
    inputs = clip_processor(images=image, return_tensors="pt").to(device)
    with torch.no_grad():
        embedding = clip_model.get_image_features(**inputs)
    return embedding.cpu().numpy().astype("float32")

def preprocess_text(text):
    """Preprocesar texto para obtener embeddings con CLIP"""
    inputs = clip_processor(text=[text], return_tensors="pt").to(device)
    with torch.no_grad():
        embedding = clip_model.get_text_features(**inputs)
    return embedding.cpu().numpy().astype("float32")

def retrieve_similar_items(query_image=None, query_text=None, k=5):
    """Recuperar elementos similares del índice vectorial"""
    if query_image is not None:
        query_embedding = preprocess_image(query_image)
        distances, indices = image_index.search(query_embedding, k)
        search_type = "imagen"
        query_for_rag = "imagen subida por el usuario"
    elif query_text is not None and query_text.strip() != "":
        query_embedding = preprocess_text(query_text)
        distances, indices = text_index.search(query_embedding, k)
        search_type = "texto"
        query_for_rag = query_text
    else:
        return [], "", ""

    results = []
    descriptions = []

    for i, idx in enumerate(indices[0]):
        if idx < len(metadata):
            img_path = metadata[idx]["path"]
            description = metadata[idx]["description"]
            similarity_score = 1 / (1 + distances[0][i])

            results.append({
                'path': img_path,
                'description': description,
                'similarity': similarity_score,
                'rank': i + 1
            })
            descriptions.append(f"Producto {i+1}: {description} (Similaridad: {similarity_score:.2f})")

    return results, descriptions, search_type, query_for_rag

def generate_rag_response(contexto, query, search_type):
    """Generar respuesta usando RAG con Gemini - versión simplificada"""

    if not contexto:
        return "No se encontraron productos similares en el catálogo."

    # Prompt simplificado - descripción detallada de cada objeto
    prompt = f"""
Eres un asistente experto en moda. Solo puedes responder usando el contexto de productos provisto.
Si no existe información suficiente en el contexto, analiza las imagenes recuperadas y describelas utilizando la descripcion de cada imagen lo mejor que puedas en lenguaje natural.
Responde en español de manera clara y directa.

=== CONTEXTO ===
Productos similares encontrados en el catálogo:
{contexto}

=== QUERY ===
El usuario {"subió una imagen" if search_type == "imagen" else f'busca: "{query}"'}.

Para cada producto encontrado, describe:
- Qué tipo de objeto/prenda es
- Sus características principales (color, material, estilo)
- Cómo se ve o para qué se usa

Proporciona una descripción detallada y clara de cada producto mostrado.

=== RESPUESTA ===
"""

    try:
        # Obtener respuesta generada por Gemini
        respuesta = modelo_gemini.generate_content(prompt)
        return respuesta.text
    except Exception as e:
        return f"Error al generar respuesta: {str(e)}"

def search_and_generate(query_image, query_text, num_results=5):
    """Función principal que combina búsqueda y generación RAG"""

    # Recuperar elementos similares
    results, descriptions, search_type, query_for_rag = retrieve_similar_items(
        query_image, query_text, k=num_results
    )

    if not results:
        return [], "No se encontraron resultados para tu consulta."

    # Crear contexto para RAG
    contexto = "\n".join(descriptions)

    # Generar respuesta RAG usando tu estructura
    rag_response = generate_rag_response(contexto, query_for_rag, search_type)

    # Preparar imágenes para mostrar
    output_images = []
    for result in results:
        try:
            img = Image.open(result['path'])
            caption = f"#{result['rank']} - {result['description']}\nSimilitud: {result['similarity']:.1%}"
            output_images.append((img, caption))
        except Exception as e:
            print(f"Error cargando imagen {result['path']}: {e}")

    return output_images, rag_response

def create_gradio_interface():
    """Crear interfaz Gradio"""

    with gr.Blocks(title="Sistema RAG Fashion", theme=gr.themes.Soft()) as interface:

        gr.Markdown("""
        # 🛍️ Sistema RAG Fashion

        **Búsqueda inteligente de productos de moda con IA**

        - 📸 Sube una imagen para encontrar productos similares
        - 📝 Escribe lo que buscas (ej: "red leather handbag", "casual dress")

        El sistema encuentra productos similares y te da consejos de moda personalizados.
        """)

        with gr.Row():
            with gr.Column(scale=1):
                gr.Markdown("### 🔍 Búsqueda")

                query_image = gr.Image(
                    type="pil",
                    label="📸 Sube una imagen",
                    height=250
                )

                query_text = gr.Textbox(
                    label="📝 O describe lo que buscas",
                    placeholder="Ej: vestido rojo elegante, zapatos casuales...",
                    lines=2
                )

                num_results = gr.Slider(
                    minimum=3,
                    maximum=8,
                    value=5,
                    step=1,
                    label="📊 Número de resultados"
                )

                search_btn = gr.Button("🔍 Buscar", variant="primary", size="lg")

                # Botones de ejemplo
                gr.Markdown("### 💡 Ejemplos:")
                with gr.Row():
                    ex1 = gr.Button("Bolso rojo", size="sm")
                    ex2 = gr.Button("Vestido casual", size="sm")
                    ex3 = gr.Button("Zapatos negros", size="sm")

            with gr.Column(scale=2):
                gr.Markdown("### 📊 Productos Encontrados")

                results_gallery = gr.Gallery(
                    label="",
                    show_label=False,
                    columns=3,
                    rows=2,
                    height=400
                )

                gr.Markdown("### 🤖 Sistema RAG")

                rag_response = gr.Textbox(
                    label="",
                    lines=12,
                    max_lines=20,
                    show_copy_button=True,
                    show_label=False
                )

        # Conectar botones
        search_btn.click(
            fn=search_and_generate,
            inputs=[query_image, query_text, num_results],
            outputs=[results_gallery, rag_response]
        )

        # Ejemplos
        ex1.click(lambda: (None, "bolso rojo"), None, [query_image, query_text])
        ex2.click(lambda: (None, "vestido casual"), None, [query_image, query_text])
        ex3.click(lambda: (None, "zapatos negros"), None, [query_image, query_text])

    return interface

def test_rag_system():
    """Probar el sistema RAG"""
    print("🧪 Probando sistema RAG...")

    # Prueba con consulta de texto
    test_query = "red handbag"
    results, descriptions, search_type, query_rag = retrieve_similar_items(query_text=test_query, k=3)

    print(f"\n📝 Consulta: '{test_query}'")
    print(f"🔍 Productos encontrados: {len(results)}")

    if results:
        # Crear contexto
        contexto = "\n".join(descriptions)
        print(f"\n📋 Contexto generado:")
        print(contexto)

        # Generar respuesta RAG
        print(f"\n🤖 Generando respuesta con Gemini...")
        rag_response = generate_rag_response(contexto, query_rag, search_type)
        print(f"\n✅ Respuesta RAG:")
        print(rag_response)
    else:
        print("❌ No se encontraron resultados")

def main():
    """Función principal"""
    print("🚀 Iniciando Sistema RAG Fashion...")

    # Verificar archivos
    required_files = [
        f"{base_path}/clip_image_index.faiss",
        f"{base_path}/fashion_text_index.faiss",
        f"{base_path}/clip_metadata.json"
    ]

    missing = [f for f in required_files if not os.path.exists(f)]
    if missing:
        print("❌ Faltan archivos:")
        for f in missing:
            print(f"  - {f}")
        return

    print("✅ Archivos verificados")
    print("✅ Gemini configurado")
    print("🎯 Lanzando interfaz...")

    # Crear y lanzar interfaz
    interface = create_gradio_interface()
    interface.launch(share=True, debug=True)

# Ejecutar
if __name__ == "__main__":
    # Descomentar para probar primero
    # test_rag_system()

    # Lanzar interfaz
    main()
```
